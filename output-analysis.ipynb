```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('output.csv', header=None, delimiter=';')
```

```python
# interfaces = []
interfaces = [0, 3.5e-3, 4.1e-3, 5.2e-3]
```

```python
fig, axs = plt.subplots(2, 3)
plt.subplots_adjust(top=1.2, right=2.0)
axs[0][0].plot(data[1], data[2], 'bo')
axs[0][0].set_title('|Ex|')
axs[0][1].plot(data[1], data[4], 'bo')
axs[0][1].set_title('|Ey|')
axs[0][2].plot(data[1], data[6], 'bo')
axs[0][2].set_title('|Ez|')
axs[1][0].plot(data[1], data[3], 'bo')
axs[1][0].set_title('arg(Ex)')
axs[1][1].plot(data[1], data[5], 'bo')
axs[1][1].set_title('arg(Ey)')
axs[1][2].plot(data[1], data[7], 'bo')
axs[1][2].set_title('arg(Ez)')

for i in range(2):
    for j in range(3):
        for interface in interfaces:
            axs[i][j].axvline(x=interface, c='k')
```
