$$
n=\frac{c_0}{c}=\frac{\sqrt{\varepsilon\mu}}{\sqrt{\varepsilon_0\mu_0}}=\sqrt{\frac{\varepsilon}{\varepsilon_0}\frac{\mu}{\mu_0}}=\sqrt{\varepsilon_r\mu_r}
$$
$$
\mu_r=1\implies n=\sqrt{\varepsilon_r}\iff\varepsilon_r=n^2
$$

```python
# convert refraction indices into relative epsilon assuming relative mu = 1

refraction_indices = [
    1.0,
    1.331 - 1.257e-7j,
    1.4 - 5.479e-6j,
    1.511 - 1.096e-8j,
    1.093 - 0.05j
]

for refraction_index in refraction_indices:
    relative_epsilon = refraction_index**2
    print(f'{relative_epsilon.real}, {relative_epsilon.imag}, 1.0, 0.0')
```
